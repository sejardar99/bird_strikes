```python
import pandas as pd
from geopy.geocoders import Nominatim
from time import sleep
```

```python
#Loading the Bird Strikes data as Pandas dataframe

file_path = r"F:\UM Internship\Bird Strikes Visualization\Bird Strikes data cleaned.xlsx"
dataset = pd.read_excel(file_path)
```

```python
#Grouping by airports with SUM aggregate to get the total no. of bird strikes for each unique airport

airport_grouped = dataset.groupby('Airport: Name').agg({'Wildlife: Number Struck Actual':'sum'})
```

```python
airport_grouped_reset = airport_grouped.reset_index()
```

```python
#Sorting the airports in descending order of no. of bird strikes 

airport_group_sort = airport_grouped_reset.sort_values(by='Wildlife: Number Struck Actual', ascending=False)
```

```python
airport_group_sort
```

```python
airport_group_sort.value_counts('Wildlife: Number Struck Actual')
```

```python
#Filtering the data to get airports with 10 and above bird strikes

TenAndAboveStrikes = airport_group_sort[airport_group_sort['Wildlife: Number Struck Actual'] >= 10]
```

```python
TenAndAboveStrikes
```

```python
TenAndAboveStrikes = TenAndAboveStrikes.reset_index(drop=True)
```

```python
TenAndAboveStrikes
```

```python
#Function to perform some data cleaning with string function 'replace', to prepare the data for geocoding 

def edit_airport(airport):
    airport = airport.replace('INTL', 'INTERNATIONAL').replace('ARPT', 'AIRPORT')
    if 'AIRPORT' not in airport:
        airport += ' AIRPORT'
    return airport
```

```python
#Applying the function on the filtered data

TenAndAboveStrikes['Airport: Name'] = TenAndAboveStrikes['Airport: Name'].apply(edit_airport)
```

```python
TenAndAboveStrikes
```

```python
#Function to perform geocoding using the Nominatim geocoder of geopy library.
#This function will take an airport name as input and return its coordinates (latitude and longitude) as output.
#Using sleep function of time library, a time delay of 1 second has been put after every geocoding API call.

#For airports that could not be geocoded, 'NA' values are returned.
#Names of such airports have been further cleaned by cross-checking from the internet;
#then the data has been updated with the corrected airport names, and again geocoded with this function successfully


def get_coordinates(airports):
    geolocator = Nominatim(user_agent='')
    airport_lat, airport_long = [], []
    for a in airports:
        location = geolocator.geocode(a)
        if location:
            airport_lat.append(location.latitude)
            airport_long.append(location.longitude)
        else:
            airport_lat.append('NA')
            airport_long.append('NA')
        sleep(1)
    return (airport_lat, airport_long)
```

```python
TenAndAboveStrikes['Wildlife: Number Struck Actual'].unique()
```

```python
#Filtering the '10 and above strikes' dataframe to get airports with 1000 and above bird strikes

ThousandAndAboveStrikes = TenAndAboveStrikes[TenAndAboveStrikes['Wildlife: Number Struck Actual'] >= 1000]
```

```python
ThousandAndAboveStrikes
```

```python
#Cleaning airport names for successful geocoding

ThousandAndAboveStrikes.loc[2, 'Airport: Name'] = ThousandAndAboveStrikes.loc[2, 'Airport: Name'].replace('NY', 'NEW YORK')
ThousandAndAboveStrikes.loc[9, 'Airport: Name'] = ThousandAndAboveStrikes.loc[9, 'Airport: Name'].replace('WASH', 'WASHINGTON')
```

```python
ThousandAndAboveStrikes
```

```python
#Getting the airport coordinates for '1000 and above strikes' data using the geocoding function

lat_1000, long_1000 = get_coordinates(ThousandAndAboveStrikes['Airport: Name'])
```

```python
#Adding the obtained coordinates to the dataframe by creating new columns for latitudes and longitudes

ThousandAndAboveStrikes.loc[:, 'latitude'] = lat_1000
ThousandAndAboveStrikes.loc[:, 'longitude'] = long_1000
```
